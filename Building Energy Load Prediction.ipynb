Scaling is done to bring all features to same level of magnitudes. Without scaling the features with high magnitudes will have more weight in the best fit.

```python
from sklearn.preprocessing import StandardScaler #Standardises the data to a range in which the mean is equal to zero and the standard deviation is 1.
from sklearn.preprocessing import MinMaxScaler #normalizes the data and brings the values between 0(lowest value) and 1(highest value)
from sklearn.preprocessing import RobustScaler #Standardizes the data but is more robust to outliers because it only scales the data according to Interquartile Range

#Generate the scaler
sc1=StandardScaler()
sc2=MinMaxScaler()
sc3=RobustScaler()
```

```python
#Scaling the input data
x1 = sc1.fit_transform(knmi_updated)
x2 = sc2.fit_transform(knmi_updated)
x3 = sc3.fit_transform(knmi_updated)
```

# RNN Implementation

```python
import numpy as np

# assume your features are stored in X and your energy consumption is stored in Y
# reshape X and Y into a 3D array of shape (samples, timesteps, features)
# where timesteps = 3 and features = number of features in X
timesteps = 3
X_reshaped = np.array([x1[i:i+timesteps,:] for i in range(len(x1)-timesteps+1)])
Y_reshaped = Energy[timesteps-1:]

# X_reshaped now has shape (samples, timesteps, features)
# Y_reshaped now has shape (samples, output_dim)
output_dim = 1

# split X_reshaped and Y_reshaped into training and validation sets
train_size = int(len(X_reshaped)*0.8)
X_train, X_val = X_reshaped[:train_size,:,:], X_reshaped[train_size:,:,:]
Y_train, Y_val = Y_reshaped[:train_size].to_numpy().flatten(), Y_reshaped[train_size:].to_numpy().flatten()

# create a sequential model
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN

from keras.layers import Dropout

model_RNN = Sequential()
model_RNN.add(SimpleRNN(units=64, input_shape=(timesteps, X_train.shape[2])))
model_RNN.add(Dropout(0.2)) # Add dropout layer with 0.2 dropout rate
model_RNN.add(Dense(units=output_dim))

model_RNN.summary()

model_RNN.compile(loss='mse', optimizer='adam')

# Add early stopping to prevent overfitting
from keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', patience=3)

historyRNN = model_RNN.fit(X_train, Y_train, validation_data=(X_val, Y_val), epochs=20, batch_size=32)
```

```python
print(r2_score(Y_train, model_RNN.predict(X_train)))
print(r2_score(Y_val, model_RNN.predict(X_val)))
print(mean_squared_error(Y_val, model_RNN.predict(X_val)))
```

```python
import matplotlib.pyplot as plt

loss_train = historyRNN.history['loss']
loss_val = historyRNN.history['val_loss']
EPOCHS = range(0,20)
plt.plot(EPOCHS, loss_train, 'g', label='Training loss')
plt.plot(EPOCHS, loss_val, 'b', label='validation loss')
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()


# Calculate the standard deviation of the training and validation losses for each epoch
train_loss_std = np.std(loss_train)
val_loss_std = np.std(loss_val)
# Plot the standard deviation of training and validation losses with respect to epoch
# plt.plot(EPOCHS, train_losses_std, label='Training Loss')
# plt.plot(EPOCHS, val_losses_std, label='Validation Loss')
plt.fill_between(EPOCHS, loss_train - train_loss_std, loss_train + train_loss_std, alpha=0.1, color='blue', label = 'Std. for Training Loss')
plt.fill_between(EPOCHS, loss_val - val_loss_std, loss_val + val_loss_std, alpha=0.1, color='green', label = 'Std. for Validation Loss')

plt.title('Standard Deviation of Losses with Epoch')
plt.xlabel('Epoch')
plt.ylabel('Standard Deviation')
plt.legend()
plt.show()
```
